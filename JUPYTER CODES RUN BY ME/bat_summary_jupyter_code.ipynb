```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
 data = json.load(f)
data[0]['matchSummary']
```

```python
match_result=pd.DataFrame(data[0]['matchSummary'])
match_result.head()
```

```python
match_result.rename({'scorecard':'Match_id'},axis=1,inplace='true')
```

```python
with open('t20_json_files/t20_wc_batting_summary.json') as g:
    batting_summary=json.load(g)
len(batting_summary)
da=pd.DataFrame(batting_summary[44]['battingSummary'])
```

```python
teams_dict_func = {}
for index,row in match_result.iterrows():
    key1 = row['team1']+ " Vs " +row['team2']
    key2 = row['team2']+ " Vs " +row['team1']
    teams_dict_func[key1] = row['Match_id']
    teams_dict_func[key2] = row['Match_id']
teams_dict_func
```

```python
da.head()
```

```python
all_records=[]
for rec in batting_summary:
   all_records.extend(rec["battingSummary"])
data_record=pd.DataFrame(all_records)
data_record.head()
```

```python
data_record.tail()
```

```python
data_record["out/not_out"]=data_record.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
```

```python
data_record.drop(columns=["dismissal"],inplace=True)
```

```python
data_record["Match_id"]=data_record["match"].map(teams_dict_func)
data_record.head()
```
