```python
with open('t20_json_files/t20_wc_match_results.json') as p:
    data=json.load(p)
len(data)    
```

```python
df_matchSummary=pd.DataFrame(data[0]['matchSummary'])
df_matchSummary.head()
```

```python
df_matchSummary.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df_matchSummary.head()
```

```python
dict_teamvs={}
for index,row in df_matchSummary.iterrows():
    key1=row['team1'] +" Vs "+row['team2']
    key2=row['team2'] +" Vs "+row['team1']
    dict_teamvs[key1]=row['match_id']
    dict_teamvs[key2]=row['match_id']
dict_teamvs
 
```

```python
df_matchSummary.tail()
```

```python
bowling_summary.head(1)
```

```python
bowling_summary["match_id"]=bowling_summary["match"].map(dict_teamvs)
bowling_summary
```

```python
bowling_summary.to_csv('t20_json_files/bowling_summary.csv', index=False)
```
